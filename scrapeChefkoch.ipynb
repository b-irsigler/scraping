```python
"""Scrape Chefkoch.de to get data of ingredients of recipes used in different cuisines.
Subsequently, scrape de.wikipedia.org for a list of spices. 
Finally, create data of spices used in different cuisines.

by Bernhard Irsigler"""
```

```python
#sources
#https://www.youtube.com/watch?v=XQgXKtPSzUI
```

```python
from urllib import request
import bs4
```

```python
#the full content is devided in subpages with 30 recipes each
recipesPerPage = 30

#we iterate over subpages until no recipes appear anymore
i_subpage = 0
while recipesPerPage > 0:
    
    #create path of subpage
    path = 'https://www.chefkoch.de/rs/s{}g115/Mexikanische-Rezepte.html'.format(30*i_subpage)
    
    #open subpage
    resp = request.urlopen(path)
    
    #and read its content
    data = resp.read()
    
    #create BeautifulSoup object
    data_soup = bs4.BeautifulSoup(data,'html.parser')
    
    #scrape a list of recipe names
    recipes = data_soup.findAll('a',{'class':'ds-mb ds-mb-row ds-card rsel-recipe bi-recipe-item'})
    
    #get the number of reipes of subpage
    recipesPerPage = len(recipes)
    
    for i_recipe in range(recipesPerPage):
        print(str(30*i_subpage+i_recipe) + ', ' + str(recipes[i_recipe]['data-vars-recipe-title']))
        
    #increase counter for subpages
    i_subpage += 1
    
```
